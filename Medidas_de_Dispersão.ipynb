## 6.6. Exercício - Medidas de Dispersão

Utilizando média e medidas de dispersão, faça o seguinte para a base de nascidos vivos de SE em 2017:

1.   Compare as idades das mães que fizeram cesárea e das mães que fizeram parto normal
2.   Compare o peso da criança ao nascer das mães que fizeram cesárea e das mães que fizeram parto normal
3.   Compare o peso da criança ao nascer para as diferentes categorias de consultas pré-natal
4.   Compare quantidade de filhos vivos e mortos em relação ao tipo de gravidez (única, dupla, tripla ou mais)
5.   Compare a quantidade total de gravidezes (filhos vivos + filhos mortos) em relação a escolaridade

```python
#Transformar texto em número
sinasc['IDADEMAE'] = pd.to_numeric(sinasc['IDADEMAE'])
sinasc['PESO'] = pd.to_numeric(sinasc['PESO'])
sinasc['QTDFILVIVO'] = pd.to_numeric(sinasc['QTDFILVIVO'])
sinasc['QTDFILMORT'] = pd.to_numeric(sinasc['QTDFILMORT'])
```

```python
#1. Compare as idades das mães que fizeram cesárea e das mães que fizeram parto normal

partoNormal1 = sinasc[(sinasc['PARTO'] == '1')]
partoCesaria2 = sinasc[(sinasc['PARTO'] == '2')]

#Média
mediaIDADEMAEpartoNormal = partoNormal1['IDADEMAE'].mean()
mediaIDADEMAEpartoCesaria = partoCesaria2['IDADEMAE'].mean()

#Desvio Médio
dmNormal = partoNormal1['IDADEMAE'].mad()
dmCesaria = partoCesaria2['IDADEMAE'].mad()

#Variança
varIDADEMAEpartoNormal = partoNormal1['IDADEMAE'].var()
varIDADEMAEpartoCesaria = partoCesaria2['IDADEMAE'].var()

#Desvio padrão
dpIDADEMAEpartoNormal = partoNormal1['IDADEMAE'].std()
dpIDADEMAEpartoCesaria = partoCesaria2['IDADEMAE'].std()

#Amplitude Interquartil
qIDADEMAEpartoNormal = partoNormal1['IDADEMAE'].quantile([0.25,0.75]).values
qIDADEMAEpartoCesaria = partoCesaria2['IDADEMAE'].quantile([0.25,0.75]).values
iqrIDADEMAEpartoNormal = qIDADEMAEpartoNormal[1] - qIDADEMAEpartoNormal[0]
iqrIDADEMAEpartoCesaria = qIDADEMAEpartoCesaria[1] - qIDADEMAEpartoCesaria[0]

#Coeficiente de variação
cvIDADEMAEpartoNormal = dpIDADEMAEpartoNormal*100/mediaIDADEMAEpartoNormal
cvIDADEMAEpartoCesaria = dpIDADEMAEpartoCesaria*100/mediaIDADEMAEpartoCesaria

#Organizar em um dataframe
resultadoAgrupado = pd.DataFrame({
    'Parto':['Normal','Cesárea'],
    'Média':[mediaIDADEMAEpartoNormal,mediaIDADEMAEpartoCesaria],
    'Desvio Médio':[dmNormal,dmCesaria],
    'Variança':[varIDADEMAEpartoNormal,varIDADEMAEpartoCesaria],
    'Desvio Padrão':[dpIDADEMAEpartoNormal,dpIDADEMAEpartoCesaria],
    'Amplitude Interquartil':[iqrIDADEMAEpartoNormal,iqrIDADEMAEpartoCesaria],
    'Coeficiente de Variação':[cvIDADEMAEpartoNormal,cvIDADEMAEpartoCesaria]}).set_index('Parto')
resultadoAgrupado
```

```python
#2. Compare o peso da criança ao nascer das mães que fizeram cesárea e das mães que fizeram parto normal

#Média
mediaPESOpartoNormal = partoNormal1['PESO'].mean()
mediaPESOpartoCesaria = partoCesaria2['PESO'].mean()

#Desvio Médio
dmPesoNormal = partoNormal1['PESO'].mad()
dmPesoCesaria = partoCesaria2['PESO'].mad()

#Variança
varPESOpartoNormal = partoNormal1['PESO'].var()
varPESOpartoCesaria = partoCesaria2['PESO'].var()

#Desvio padrão
dpPESOpartoNormal = partoNormal1['PESO'].std()
dpPESOpartoCesaria = partoCesaria2['PESO'].std()

#Amplitude Interquartil
qPESOpartoNormal = partoNormal1['PESO'].quantile([0.25,0.75]).values
qPESOpartoCesaria = partoCesaria2['PESO'].quantile([0.25,0.75]).values
iqrPESOpartoNormal = qPESOpartoNormal[1] - qPESOpartoNormal[0]
iqrPESOpartoCesaria = qPESOpartoCesaria[1] - qPESOpartoCesaria[0]

#Coeficiente de variação
cvPESOpartoNormal = dpPESOpartoNormal*100/mediaPESOpartoNormal
cvPESOpartoCesaria = dpPESOpartoCesaria*100/mediaPESOpartoCesaria

#Organizar em um dataframe
resultadoAgrupado = pd.DataFrame({
    'Parto':['Normal','Cesárea'],
    'Média':[mediaPESOpartoNormal,mediaPESOpartoCesaria],
    'Desvio Médio':[dmPesoNormal,dmPesoCesaria],
    'Variança':[varPESOpartoNormal,varPESOpartoCesaria],
    'Desvio Padrão':[dpPESOpartoNormal,dpPESOpartoCesaria],
    'Amplitude Interquartil':[iqrPESOpartoNormal,iqrPESOpartoCesaria],
    'Coeficiente de Variação':[cvPESOpartoNormal,cvPESOpartoCesaria]}).set_index('Parto')
resultadoAgrupado
```

```python
#3. Compare o peso da criança ao nascer para as diferentes categorias de consultas pré-natal

preNatal_0 = sinasc[(sinasc['CONSULTAS'] == '1')]
preNatal_1_3 = sinasc[(sinasc['CONSULTAS'] == '2')]
preNatal_4_6 = sinasc[(sinasc['CONSULTAS'] == '3')]
preNatal_7 = sinasc[(sinasc['CONSULTAS'] == '4')]

#Média
mediaPreNatal_0 = preNatal_0['PESO'].mean()
mediaPreNatal_1_3 = preNatal_1_3['PESO'].mean()
mediaPreNatal_4_6 = preNatal_4_6['PESO'].mean()
mediaPreNatal_7 = preNatal_7['PESO'].mean()

#Desvio Médio
dmPreNatal_0 = preNatal_0['PESO'].mad()
dmPreNatal_1_3 = preNatal_1_3['PESO'].mad()
dmPreNatal_4_6 = preNatal_4_6['PESO'].mad()
dmPreNatal_7 = preNatal_7['PESO'].mad()

#Variança
varPreNatal_0 = preNatal_0['PESO'].var()
varPreNatal_1_3 = preNatal_1_3['PESO'].var()
varPreNatal_4_6 = preNatal_4_6['PESO'].var()
varPreNatal_7 = preNatal_7['PESO'].var()

#Desvio padrão
dpPreNatal_0 = preNatal_0['PESO'].std()
dpPreNatal_1_3 = preNatal_1_3['PESO'].std()
dpPreNatal_4_6 = preNatal_4_6['PESO'].std()
dpPreNatal_7 = preNatal_7['PESO'].std()

#Amplitude Interquartil
qPreNatal_0 = preNatal_0['PESO'].quantile([0.25,0.75]).values
qPreNatal_1_3 = preNatal_1_3['PESO'].quantile([0.25,0.75]).values
qPreNatal_4_6 = preNatal_4_6['PESO'].quantile([0.25,0.75]).values
qPreNatal_7 = preNatal_7['PESO'].quantile([0.25,0.75]).values

iqrPreNatal_0 = qPreNatal_0[1] - qPreNatal_0[0]
iqrPreNatal_1_3 = qPreNatal_1_3[1] - qPreNatal_1_3[0]
iqrPreNatal_4_6 = qPreNatal_4_6[1] - qPreNatal_4_6[0]
iqrPreNatal_7 = qPreNatal_7[1] - qPreNatal_7[0]

#Coeficiente de variação
cvPreNatal_0 = dpPreNatal_0*100/mediaPreNatal_0
cvPreNatal_1_3 = dpPreNatal_1_3*100/mediaPreNatal_1_3
cvPreNatal_4_6 = dpPreNatal_4_6*100/mediaPreNatal_4_6
cvPreNatal_7 = dpPreNatal_7*100/mediaPreNatal_7

#Organizar em um dataframe
resultadoAgrupado = pd.DataFrame({
    'Nro Consultas':['0', '1,2 ou 3', '4, 5 ou 6', '7 +'],
    'Média':[mediaPreNatal_0, mediaPreNatal_1_3, mediaPreNatal_4_6, mediaPreNatal_7], 
    'Desvio Médio':[dmPreNatal_0, dmPreNatal_1_3, dmPreNatal_4_6, dmPreNatal_7],
    'Variança':[varPreNatal_0, varPreNatal_1_3, varPreNatal_4_6, varPreNatal_7],
    'Desvio Padrão':[dpPreNatal_0, dpPreNatal_1_3, dpPreNatal_4_6, dpPreNatal_7], 
    'Amplitude Interquartil':[iqrPreNatal_0, iqrPreNatal_1_3, iqrPreNatal_4_6, iqrPreNatal_7],
    'Coeficiente de Variação':[cvPreNatal_0, cvPreNatal_1_3, cvPreNatal_4_6, cvPreNatal_7]} ).set_index('Nro Consultas')
resultadoAgrupado
```

```python
#4. Compare quantidade de filhos vivos e mortos em relação ao tipo de gravidez (única, dupla, tripla ou mais)
#Filhos VIVOS

gravidezUnico= sinasc[(sinasc['GRAVIDEZ'] == '1')]
gravidezGemeas= sinasc[(sinasc['GRAVIDEZ'] == '2')]
gravidezTriplaMais= sinasc[(sinasc['GRAVIDEZ'] == '3')]

#Média
mediaGravidezUnico = gravidezUnico['QTDFILVIVO'].mean()
mediaGravidezGemeas = gravidezGemeas['QTDFILVIVO'].mean()
mediaGravidezTriplaMais = gravidezTriplaMais['QTDFILVIVO'].mean()

#Desvio Médio
dmGravidezUnico = gravidezUnico['QTDFILVIVO'].mad()
dmGravidezGemeas = gravidezGemeas['QTDFILVIVO'].mad()
dmGravidezTriplaMais = gravidezTriplaMais['QTDFILVIVO'].mad()

#Variança
varGravidezUnico = gravidezUnico['QTDFILVIVO'].var()
varGravidezGemeas = gravidezGemeas['QTDFILVIVO'].var()
varGravidezTriplaMais = gravidezTriplaMais['QTDFILVIVO'].var()

#Desvio padrão
dpGravidezUnico = gravidezUnico['QTDFILVIVO'].std()
dpGravidezGemeas = gravidezGemeas['QTDFILVIVO'].std()
dpGravidezTriplaMais = gravidezTriplaMais['QTDFILVIVO'].std()

#Amplitude Interquartil
qGravidezUnico = gravidezUnico['QTDFILVIVO'].quantile([0.25, 0.75]).values
qGravidezGemeas = gravidezGemeas['QTDFILVIVO'].quantile([0.25, 0.75]).values
qGravidezTriplaMais = gravidezTriplaMais['QTDFILVIVO'].quantile([0.25, 0.75]).values

iqrGravidezUnico = qGravidezUnico[1] - qGravidezUnico[0]
iqrGravidezGemeas = qGravidezGemeas[1] - qGravidezGemeas[0]
iqrGravidezTriplaMais = qGravidezTriplaMais[1] - qGravidezTriplaMais[0]

#Coeficiente de variação
cvGravidezUnico = dpGravidezUnico*100/mediaGravidezUnico
cvGravidezGemeas = dpGravidezGemeas*100/mediaGravidezGemeas
cvGravidezTriplaMais = dpGravidezTriplaMais*100/mediaGravidezTriplaMais

#Organizar em um dataframe
resultadoAgrupado = pd.DataFrame({
    'Tipo Gravidez':['Normal (Um)', 'Gemeos', 'Trigemeos ou mais'],
    'Média':[mediaGravidezUnico, mediaGravidezGemeas, mediaGravidezTriplaMais], 
    'Desvio Médio':[dmGravidezUnico, dmGravidezGemeas, dmGravidezTriplaMais],
    'Variança':[varGravidezUnico, varGravidezGemeas, varGravidezTriplaMais],
    'Desvio Padrão':[dpGravidezUnico, dpGravidezGemeas, dpGravidezTriplaMais], 
    'Amplitude Interquartil':[iqrGravidezUnico, iqrGravidezGemeas, iqrGravidezTriplaMais],
    'Coeficiente de Variação':[cvGravidezUnico, cvGravidezGemeas, cvGravidezTriplaMais]} ).set_index('Tipo Gravidez')
resultadoAgrupado
```

```python
#4. Compare quantidade de filhos vivos e mortos em relação ao tipo de gravidez (única, dupla, tripla ou mais)
#Filhos MORTOS

gravidezUnicoMorto = sinasc[(sinasc['GRAVIDEZ'] == '1')]
gravidezGemeasMorto = sinasc[(sinasc['GRAVIDEZ'] == '2')]
gravidezTriplaMaisMorto = sinasc[(sinasc['GRAVIDEZ'] == '3')]

#Média
mediaGravidezUnicoMorto = gravidezUnicoMorto['QTDFILMORT'].mean()
mediaGravidezGemeasMorto = gravidezGemeasMorto['QTDFILMORT'].mean()
mediaGravidezTriplaMaisMorto = gravidezTriplaMaisMorto['QTDFILMORT'].mean()

#Desvio Médio
dmGravidezUnicoMorto = gravidezUnicoMorto['QTDFILMORT'].mad()
dmGravidezGemeasMorto = gravidezGemeasMorto['QTDFILMORT'].mad()
dmGravidezTriplaMaisMorto = gravidezTriplaMaisMorto['QTDFILMORT'].mad()

#Variança
varGravidezUnicoMorto = gravidezUnicoMorto['QTDFILMORT'].var()
varGravidezGemeasMorto = gravidezGemeasMorto['QTDFILMORT'].var()
varGravidezTriplaMaisMorto = gravidezTriplaMaisMorto['QTDFILMORT'].var()

#Desvio padrão
dpGravidezUnicoMorto = gravidezUnicoMorto['QTDFILMORT'].std()
dpGravidezGemeasMorto = gravidezGemeasMorto['QTDFILMORT'].std()
dpGravidezTriplaMaisMorto = gravidezTriplaMaisMorto['QTDFILMORT'].std()

#Amplitude Interquartil
qGravidezUnicoMorto = gravidezUnicoMorto['QTDFILMORT'].quantile([0.25, 0.75]).values
qGravidezGemeasMorto = gravidezGemeasMorto['QTDFILMORT'].quantile([0.25, 0.75]).values
qGravidezTriplaMaisMorto = gravidezTriplaMaisMorto['QTDFILMORT'].quantile([0.25, 0.75]).values

iqrGravidezUnicoMorto = qGravidezUnicoMorto[1] - qGravidezUnicoMorto[0]
iqrGravidezGemeasMorto = qGravidezGemeasMorto[1] - qGravidezGemeasMorto[0]
iqrGravidezTriplaMaisMorto = qGravidezTriplaMaisMorto[1] - qGravidezTriplaMaisMorto[0]

#Coeficiente de variação
cvGravidezUnicoMorto = dpGravidezUnicoMorto*100/mediaGravidezUnicoMorto
cvGravidezGemeasMorto = dpGravidezGemeasMorto*100/mediaGravidezGemeasMorto
cvGravidezTriplaMaisMorto = dpGravidezTriplaMaisMorto*100/mediaGravidezTriplaMaisMorto

#Organizar em um dataframe
resultadoAgrupado = pd.DataFrame({
    'Tipo Gravidez':['Normal (Um)', 'Gemeos', 'Trigemeos ou mais'],
    'Média':[mediaGravidezUnicoMorto, mediaGravidezGemeasMorto, mediaGravidezTriplaMaisMorto], 
    'Desvio Médio':[dmGravidezUnicoMorto, dmGravidezGemeasMorto, dmGravidezTriplaMaisMorto],
    'Variança':[varGravidezUnicoMorto, varGravidezGemeasMorto, varGravidezTriplaMaisMorto],
    'Desvio Padrão':[dpGravidezUnicoMorto, dpGravidezGemeasMorto, dpGravidezTriplaMaisMorto], 
    'Amplitude Interquartil':[iqrGravidezUnicoMorto, iqrGravidezGemeasMorto, iqrGravidezTriplaMaisMorto],
    'Coeficiente de Variação':[cvGravidezUnicoMorto, cvGravidezGemeasMorto, cvGravidezTriplaMaisMorto]} ).set_index('Tipo Gravidez')
resultadoAgrupado
```

```python
#5. Compare a quantidade

semEscolaridade= sinasc[(sinasc['ESCMAE'] == '1')]
escolaridade_1_3= sinasc[(sinasc['ESCMAE'] == '2')]
escolaridade_4_7= sinasc[(sinasc['ESCMAE'] == '3')]
escolaridade_8_11= sinasc[(sinasc['ESCMAE'] == '4')]
escolaridade_12_mais= sinasc[(sinasc['ESCMAE'] == '5')]

semEscolaridade['FILHOS'] = semEscolaridade['QTDFILVIVO'] + semEscolaridade['QTDFILMORT']
escolaridade_1_3['FILHOS']  = escolaridade_1_3['QTDFILVIVO'] + escolaridade_1_3['QTDFILMORT']
escolaridade_4_7['FILHOS']  = escolaridade_4_7['QTDFILVIVO'] + escolaridade_4_7['QTDFILMORT']
escolaridade_8_11['FILHOS']  = escolaridade_8_11['QTDFILVIVO'] + escolaridade_8_11['QTDFILMORT']
escolaridade_12_mais['FILHOS']  = escolaridade_12_mais['QTDFILVIVO'] + escolaridade_12_mais['QTDFILMORT']

#Média
mediaSemEscolaridade = semEscolaridade['FILHOS'].mean()
mediaEscolaridade_1_3 = escolaridade_1_3['FILHOS'].mean()
mediaEscolaridade_4_7 = escolaridade_4_7['FILHOS'].mean()
mediaEscolaridade_8_11 = escolaridade_8_11['FILHOS'].mean()
mediaEscolaridade_12_mais = escolaridade_12_mais['FILHOS'].mean()

#Desvio Médio
dmSemEscolaridade = semEscolaridade['FILHOS'].mad()
dmEscolaridade_1_3 = escolaridade_1_3['FILHOS'].mad()
dmEscolaridade_4_7 = escolaridade_4_7['FILHOS'].mad()
dmEscolaridade_8_11 = escolaridade_8_11['FILHOS'].mad()
dmEscolaridade_12_mais = escolaridade_12_mais['FILHOS'].mad()

#Variança
varSemEscolaridade = semEscolaridade['FILHOS'].var()
varEscolaridade_1_3 = escolaridade_1_3['FILHOS'].var()
varEscolaridade_4_7 = escolaridade_4_7['FILHOS'].var()
varEscolaridade_8_11 = escolaridade_8_11['FILHOS'].var()
varEscolaridade_12_mais = escolaridade_12_mais['FILHOS'].var()

#Desvio padrão
dpSemEscolaridade = semEscolaridade['FILHOS'].std()
dpEscolaridade_1_3 = escolaridade_1_3['FILHOS'].std()
dpEscolaridade_4_7 = escolaridade_4_7['FILHOS'].std()
dpEscolaridade_8_11 = escolaridade_8_11['FILHOS'].std()
dpEscolaridade_12_mais = escolaridade_12_mais['FILHOS'].std()

#Amplitude Interquartil
qSemEscolaridade = semEscolaridade['FILHOS'].quantile([0.25,0.75]).values
qEscolaridade_1_3 = escolaridade_1_3['FILHOS'].quantile([0.25,0.75]).values
qEscolaridade_4_7 = escolaridade_4_7['FILHOS'].quantile([0.25,0.75]).values
qEscolaridade_8_11 = escolaridade_8_11['FILHOS'].quantile([0.25,0.75]).values
qEscolaridade_12_mais = escolaridade_12_mais['FILHOS'].quantile([0.25,0.75]).values

iqrSemEscolaridade = qSemEscolaridade[1] - qSemEscolaridade[0]
iqrEscolaridade_1_3 = qEscolaridade_1_3[1] - qEscolaridade_1_3[0]
iqrEscolaridade_4_7 = qEscolaridade_4_7[1] - qEscolaridade_4_7[0]
iqrEscolaridade_8_11 = qEscolaridade_8_11[1] - qEscolaridade_8_11[0]
iqrEscolaridade_12_mais = qEscolaridade_12_mais[1] - qEscolaridade_12_mais[0]

#Coeficiente de variação
cvSemEscolaridade = dpSemEscolaridade*100/mediaSemEscolaridade
cvEscolaridade_1_3 = dpEscolaridade_1_3*100/mediaEscolaridade_1_3
cvEscolaridade_4_7 = dpEscolaridade_4_7*100/mediaEscolaridade_4_7
cvEscolaridade_8_11 = dpEscolaridade_8_11*100/mediaEscolaridade_8_11
cvEscolaridade_12_mais = dpEscolaridade_12_mais*100/mediaEscolaridade_12_mais

#Organizar em um dataframe
resultadoAgrupado = pd.DataFrame({
    'Escolaridade':['Sem escolaridade', '1 a 3 anos', '4 a 7 anos', '8 a 11 anos', '12 anos ou mais'],
    'Média':[mediaSemEscolaridade, mediaEscolaridade_1_3, mediaEscolaridade_4_7, mediaEscolaridade_8_11, mediaEscolaridade_12_mais], 
    'Desvio Médio':[dmSemEscolaridade, dmEscolaridade_1_3, dmEscolaridade_4_7, dmEscolaridade_8_11, dmEscolaridade_12_mais],
    'Variança':[varSemEscolaridade, varEscolaridade_1_3, varEscolaridade_4_7, varEscolaridade_8_11, varEscolaridade_12_mais],
    'Desvio Padrão':[dpSemEscolaridade, dpEscolaridade_1_3, dpEscolaridade_4_7, dpEscolaridade_8_11, dpEscolaridade_12_mais], 
    'Amplitude Interquartil':[iqrSemEscolaridade, iqrEscolaridade_1_3, iqrEscolaridade_4_7, iqrEscolaridade_8_11, iqrEscolaridade_12_mais],
    'Coeficiente de Variação':[cvSemEscolaridade, cvEscolaridade_1_3, cvEscolaridade_4_7, cvEscolaridade_8_11, cvEscolaridade_12_mais]} ).set_index('Escolaridade')
resultadoAgrupado
```
